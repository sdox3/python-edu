#### 4. 강제중단/ 재실행
* Kernel interrupt

* 제목줄 아래에 Kernel 메뉴가 있다. 이 커널은 iPython(Jupyter Notebook)대화창 아래에서 백그라운드와 비슷하게 실행되는 일종의 운영체제와 같은 개념이다.

* kernel의 모든 메뉴는 코드를 삭제하지 않는다.
* interrupt : 실행중인 코드를 강제로 중지, 에러메시지가 뜨면서 실행이 중지가 된다.
* Restart : 실행중인 코드가 중지가 되며 재시작된다. 코드나 실행결과는 삭제가 되지 않는다.
* restart&run all : 재 시작후에 모든 셀의 코드를 위에서부터 순차적으로 실행한다.
* restart& Clear Output : 코드가 중지되며, 실행결과는 삭제가 된다.
* reconnect : 인터넷이 끊겼을 경우 다시 접속
* shutdown : 커널을 종료한다. 


* shutdown이 되었거나 인터넷 연결이 끊어 졌거나 기타 문제가 있다면 탭에 알림표시가 나타난다.

* 현재 실행중인 커널이 있는지 여부를 확인하는 방법은 첫 번째는 Home화면에서  <code>.ipynb</code> 파일의 아이콘이 초록색이면 실행중, 아니면 중단 또는 시작되느 않은 상태이다. 두번째 방법은 Running탭을 클릭하면 현재 실행중인 ipython과 터미널의 목록을 확인할 수 있다.

### E. Jupyter NoteBook의 기능

#### 1. DocString 표시

* 선언한 변수 뒤에 <code>?</code>를 붙여서 셀을 실행하는 것으로 해당 변수의 상태를 확인할 수가 있다.
* 또한 변수를 선언한 후에 <code>Shift + Tab</code>를 누르면 Tool Tip이 표시가 된다. Toop Tip에는 DocString의 일부내용이 표시된다.

* 변수 또는 함수들의 자세한 내용을 보기 위해서는 <code>help()</code>함수를 사용한다.

#### 2. Image 첨부하기

* 그림파일을 첨부하기 위해서는 <code>img tag</code>를 이용하거나 Drag&Drop를 하면된다.

#### 3. shell(명령 프롬프트)의 이용

* cmd창에서 쓰는 명령을 그대로 사용하는데 맨 앞에 <code>!</code>를 붙여서 사용하면 된다.
* ex) !pip install Pandas, !dir , !cd documents
