### 함수(Function)

함수란? 프로그램을 작성하다 보면 동일한 내용을 반복해서 작성하는 경우가 있는데 이런 반복적인 코드들을 함수로 선언하고 필요할 때 마다 재사용할 수 있도록 하기 위해서 사용한다.

파이썬은 함수(Function)라는 기능을 제공하는 데 특정용도의 코드를 한 곳에 모아 놓은 것을 의미하는 데 print(), input() 등이 파이썬에서 내장되어 있는 내장함수이고 사용자가 정의한 사용자 함수가 있다.

#### 1. 함수사용의 이점
1. 코드의 사용목적에 따라 구분할 수 있다
2. 코드의 재사용을 할 수 있다.
3. 코딩의 실수를 줄일 수 있게 된다.

#### 7. 입력되는 매개변수가 몇개인지 모르는 경우
> def 변수명<code>(*args)</code>: 실행문장 <br>
> def 변수명<code>(**kwards)</code>: 실행문장

```python
# 1. *args : 입력되는 값이 여러개 일 경우 정의하는 매개변수
# 1~100 까지 합계 숫자가 가변일 시
# args는 예약어가 아닌 변수로 처리되기 때문에 ''*임의의 변수명' 형태로 정의하면 된다.

def manysum(*args) :
    print(type(args))
    print(len(args))
    sum = 0
    for i in args :
        sum += i
    return sum
print(manysum(1,2))
print(manysum(1,2,3,4,5,6,7))
```

```python
# 실습 : 유동적으로 작동되는 계산기 함수 만들기
# 사칙연산을 하는 계산기 함수를 만들기
def calculator(operator, *args):
    if operator == '+' :
        sum = 0
        for i in args :
            sum += i
        return  sum
    elif operator == '-' :
        min = args[0]
        for i in args[1:len(args)]:
            min -= i
        return min
    elif operator == '*' :
        mul = args[0]
        for i in args[1:len(args)]:
            mul *= i
        return mul
    elif operator == '/' :
        div = args[0]
        for i in args[1:len(args)]:
            div /= i
        return div
    else :
        print('사칙연산을 입력해 주세요')
print(calculator('+',8,2,1))
print(calculator('-',8,2,1))
print(calculator('*',8,2,1))
print(calculator('/',8,2,1))
```

#### 2. *kwards

**(keyworgs)는 key word arguments의 약자이다. *args와는 다르게 별표가(*) 2개가 사용된다.
이 의미는 함수인수로서 **kwargs로 주어졌다면 key와 value 한쌍으로 입력값이 전달되면서
이 인수의 자료형은 딕셔너리 형태로 저장된다.



```python
def kwfunc(**kwargs):
    print(kwargs)

kwfunc(name='소향',age=41,address='인천')
## **kwargs는 키와 벨류를 같이 받지 않으면 에러
## kwfunc(1) ,TypeError: kwfunc() takes 0 positional arguments but 1 was given
```

```python
# 3. *args와 **kwargs을 동시에 사용할 경우
# 값만 있는 경우와 key=value전달 할 경우에는 정의된 순서에 맞게 전달해야 한다.

def mixfunc(*args, **kwargs):
    print(args, kwargs)

mixfunc(1, name ="소향")
mixfunc(1,2,3, name ="소향",age=41)
# mixfunc(name ="소향",age=41,1,2,3) SyntaxError: positional argument follows keyword argument

```

#### 11. 람다함수 (Lamda)

파이썬에서 <code>lambda함수</code>란 런타임에서 사용할 수 있는 <code>익명함수</code>이다.
lamda함수는 함수를 생성할 때 사용하는 예약어로 def와 동일하지만 def로 생성한 함수는
임의의 이름을 가져야 하지만 lamda로 생성되는 함수는 이름이 없다. 함수이름이 없기 때문에 
익명함수라고 한다.

<br>

<code>lambda 함수</code>를 사용하는 가장 큰 이유는 수행할 명령이 하나만 있는 함수와 같이 def로 정의할 정도로 복작하지 않거나 def를 사용할 수 없는 곳에서 사용된다.

> lambda의 사용법
* lambda 매개변수 1... 매개변수n: 매개변수를 이용한 한개의 실행문

```python
# 1. def로 정의할 만큼 복잡하지 않은 경우

# 일반 함수
def add(a,b) :
    return a+b
print(add(1,1))
print(add)

# lambda 함수(이름이 없는 함수)
addlambdafunction = lambda a,b : a+b
print(addlambdafunction)
print()

print(type(add))
print(type(addlambdafunction))
print()

print(addlambdafunction(1,1))
```

```python
# 2. def를 사용할 수 없는 곳에서 사용할 경우
# lambda식이 나오게 된 가장 큰 이유는 def보다 간결하게 사용할 수 있기 때문이기도 하지만
# lambda는 def를 사용할 수 없는 곳에서도 사용할 수 있다.

def sum(a,b) : return a+b
def sub(a,b) : return a-b
x = sum(1,1)
y = sub(1,1)
myList = [x,y]
print(myList)
print(type(myList))
print(myList[0], myList[1])
print()

# list와 tuple 안에는 def를 사용할 수 없기 때문에 사용하고 싶으면 lambda를 사용한다.
myList2 = [sum, sub]

print(myList2[0](1,1), myList2[1](1,1))
print(type(myList2))
print()

def sub1(a,b) : a-b
myList3 =[sum, sub1]
print(myList3)
print(myList3[0](1,1))
print(myList3[1](1,2))
print()

# lambda로 list 정의
myList4 = [lambda a,b : a+b, lambda a,b:a-b]
print(type(myList4))
print(myList4[0](1,2))
print(myList4[1](1,2))
```

### 연습문제

```python
def isodd(number):
    if(number%2 == 0):
        print('짝수입니다.')
    else :
        print('홀수입니다.')
        
isOddLambda = lambda number : print('짝수입니다.') if number%2 ==0 else print('홀수입니다')

isodd(1)
isodd(2)
isOddLambda(2)
isOddLambda(1)
```

```python
# 2. 입력된 모든 수의 평균을 계산하는 funcAvg()함수를 작성
# 단, 입력매개변수의 갯수는 정해져 있지 않다.
# funcAvg()

def funcAvg(*args) :
    a = 0
    for i in args :
        a += i
    return a/len(args)
a = funcAvg(1,2,3,4,5)
print(a)

# 매개값을 list자료형으로 전달
# list 값은 1개의 값이기 때문에 *args로 받으면 에러
# *list를 붙여서 매개변수를 넣어줘야함
b = [1,2,3,4,5]

a = funcAvg(*b)
print(a)
```

```python
# 3. 구구단을 출력하는 함수 gugudan()함수를 작성
# 입력값은 2~9단까지 한개의 임의의 단을 입력
# gugudan(9)
def gugudan(x):
    for i in range(1, 10):
        print('%d * %d = %2d' %(x, i, x*i))
        
gugudan(9)
```

```python
# 4. 리스트 [2,3,4,5,6,7,8]에서 5보다 큰 수만 리턴하는 
# 1) funcReturn()함수작성

def funcReturn(*args) :
    j =[]
    for i in args :
        if i > 5 :
            j.append(i)
    return j
a = [2,3,4,5,6,7,8]
print(funcReturn(*a))


# 2) lambda식으로 작성, 이름은 funcReturnLambda
# 리스트로 감싼 상태에서 for, if를 해서 결과값이 리스트 안에 들어간다.
funcReturnLambda = lambda a : [number for number in a if number > 5] 
print(funcReturnLambda(a))
```

#### 파이썬의 NameSpace(네임스페이스)
<img src ='./images/12.함수_function_07.png' width =400 height = 400>

<code>NameSpace(네임스페이스)</code>이란 프로그래밍언어에서 특정한 객체를 이름(name)에 따라 구분할 수 있는 범위를 말한다. 파이썬 내부에는 모든 것은 객체로 구성되면 이들 각각은 특정이름과 맵핑관계를 갖게 되는데 이 매핑을 포함하고 있는 공간을 네임스페이스라고 한다.

네임스페이스가 필요한 이유는 다음과 같이 프로그래밍을 수행하다 보면 모든 변수이름과 함수이름을 정하는 것이 중요한데 이를 모두 겹치지 않게 정하는 것은 번거롭고 복잡한 일이다.

> 따라서, 프로그래밍언어에서는 네임스페이스라는 개념을 도입하여, 특정한 하나의 이름이 통용될 수 있도록 행위를 제한한다. 즉 소속된 네임스페이스가 다르다면 동일 이름이라도 다른 객체로 관리된다.

파이썬의 네임스페이스는 3가지로 분류된다.
>* 지역네임스페이스 : <code>함수 및 메서드</code>별로 존재하며, 함수내의 지역변수들이 소속된다.
* 전역 네임 스페이스 : <code>모듈</code>별로 존재하며, 모듈 전체에서 통용될 수 있는 변수들이 소속된다.
* 빌트인네임스페이스 : <code>기본 내장함수 및 기본예외</code>들의 이름이 소속된다. 파이썬으로 작성된 모든 코드 범위가 포함된다.

#### 파이썬의 네임스페이스는 다음과 같은 특징들이 있다.
>* 네임스페이스는 딕셔너리형태로 구현된다.
* 모든 이름 자체는 문자열로 되어 있고 각각은 해당 네임스페이스의 범위에서 실제 객체를 가르킨다.
* 이름과 실제 객체 사이의 매핑은 가변적(mutable)이기 때문에 실행동안 새로운 이름을 추가할 수 있다
* 다만, 빌드인네임스페이스는 함부로 추가하거나 삭제할 수 없다.

```python
# 네임스페이스 
# 파이썬에서는 변수는 네임스페이스에 저장된다.
# 네임스페이스를 확인하려면 globals,locals()를 사용하면 현재의 네임스페이스를 딕셔너리 
# 형태로 출력한다. 또한 주피터노트북에서는 매직명령어 %who로 확인 가능하다.

a =3 

%who
print(locals())
```

```python
# 함수는 전역과는 별개의 네임스페이스를 가진다.
# 주피터에서는 프롬프트 화면의 string도 표시가 되어서 지역변수가 보이지만
# 기본적으로는 globals를 통해서는 지역변수를 확인할 수 없다.
a = 3
def func() :
    a =100
    b = 'python'
func()
print(globals())
```

```python
print(locals())

# 함수안에서 변수 a의 값을 변경하고 b라는 새로운 변수를 생성하면 globals의 네임스페이스에는
# 변경된 값과 새로 할당된 변수이름이 보이지 않는다.
# 함수는 함수안에 별도로 구분되 네임스페이스 공간을 가진다.
```

```python
def func1() :
    a =100
    b = 'python'
    print('함수 내부 namespace', locals())
    print('함수 외부 namespace', globals())
func1()
```

#### 중첩함수 즉, 함수안에 함수 만들기


```python
def printHello():
    hello = 'hello, Python'
    def printMessage():
        print(hello)
    printMessage()
printHello()

# printHello를 호출해 printMessage를 호출한 형태
# 직접 외부에서 printMessage를 호출하면 접근이 불가하기 때문에 에러가 발생함
```

#### 클로저 사용하기
함수를 클로저형태로 만드는 방법은 함수 바깥쪽에 있는 지역변수를 사용하여 내부함수에 사용하는 방법

###### 클로저 개념도

```python
# 클로저 사용하기

def cale():
    a = 3 
    b = 5 
    def add(x):
        """클로저 개념잡기"""
        return a*x+b #외부함수의 변수(a,b)를 이용한 계산
    # 함수 add()를 만든 뒤에 이 함수를 직접 호출하지 않고 return 문을 이용해 함수자체를
    # 반환 함수를 반환할 때는 함수이름만 반환해야 한다. ()를 붙여서는 안된다.
    return add # add의 결과값이 아닌 add()자체를 반환

# cale(100) #제외처리는 아니지만 우리가 원하는 값 add가 나오지 않는다 x 정의한 경우

# cale()는 add함수 자체를 리턴을 한다.
# returnAdd에는 cale()가 있는게 아니라 add가 들어있는것
# 따라서, returnAdd(1)은 add(1)과 동일하다고 볼 수 있다.
returnAdd = cale()
print(returnAdd(200))
returnAdd?

# 이와 같이 함수를 둘러싼 환경(지역변수, 지역함수 등)을 계속 유지하다가 함수를 호출할 때
# 다시 꺼내서 사용하는 함수를 클로저라고 한다.
# 위의 경우 cale가 클로저
```
