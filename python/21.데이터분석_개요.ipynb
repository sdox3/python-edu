```python
# pandas 그래프 그리기
s.plot()
```

```python
# 3. numpy : 2차원 자료 생성
# numpy로 랜덤하게 10행3열의 2차원 행렬을 생성
data_set = np.random.rand(10,3)
print(data_set)
print(type(data_set))
```

```python
# 4. pandas : 2차원 자료구조 (dataframe)

df = pd.DataFrame(data_set)
print(df)
print(type(df))

df1 = pd.DataFrame(data= data_set, columns=['2017','2018','2019'])
```

```python
# pandas로 2차원 그래프 그리기(line)

df1.plot()
```

```python
df1.plot(kind = 'bar')
df1.plot(kind = 'area', stacked=False)
```

```python
# pie chart
# 5개의 난수를 발생, Seres자료형의 인덱스
# name = 그래프 제목
ser = pd.Series(data = np.random.rand(5),
                index=['2015','2016','2017','2018','2019'],
                name='Sales Amount')
print(ser)

# 시각화
# autopct - 데이터의 문자열 포멧을 지정 %d, %f 등 문자열 포맷으로 지정
# %.2%% - 소수점 2번째 까지 표현 %는 포맷시작과 같아서 두번 사용해 문자열임을 표현
# figsize(x,y) = x,y를 기준으로 차트의 사이즈
# fontsize = 폰트크기
ser.plot(kind ='pie', autopct='%.2f%%',figsize =(10,5), fontsize = (20))
```
