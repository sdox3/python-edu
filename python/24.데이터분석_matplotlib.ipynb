##### 그래프의 범위지정

그래프를 보면 몇몇점들을 그래프 변위 경계선에 있어 잘 보이지 않는 경우가 있을 수 있다. 이럴 
경우 그래프의 범위를 수동으로 지정할 경우 `xlim`명령과  `ylim`명령을 사용한다. 이 명령은
그래프의 범위가 되는 x축과 y축의 최소값과 최대값을 지정한다.

```python
# 그래프 x,y축 범위를 지정
plt.title('스타일이 적용된 그래프 ')
plt.plot([10,20,30,40],[2,4,6,8],c='b',lw=5,ls='--',marker ='o',ms = 15
        ,mec = 'g', mew = 5, mfc = 'r')
plt.xlim(0,50)
plt.ylim(-10,30)
plt.grid() # default True
plt.show()
```

#### 틱(tick)설정

그래프에서 축상의 위치표시 지점을 `틱(tick)` 이라고 한다.이 틱에 쓰여진 숫자 혹은 글자를 
`틱라벨(tick label)`이라고 한다. 틱의 위차나 틱라벨은 matplotlib이 자동으로 정해주지만 수동으로 설정하고자 할 경우에는 `xticks, yticks` 명령을 사용한다.

```python
# x, y의 틱라벨 설정하기
# np.linspace 시작, 끝 까지 점점 증가하는 숫자를 제공
X = np.linspace(-np.pi,np.pi,256)
X
Y = np.cos(X)
plt.title('x축과 y축의 Tick Label 설정')
plt.plot(X, Y)
plt.xticks([-np.pi, -np.pi/2, np.pi/2, np.pi])
plt.yticks([-1,0,1])
plt.grid()
plt.show()
```

#### figure객체

모든 그래프에는 `Figure객체`, 정식으로는 `Matplotlib.figure.Figure`클래스 객체가 포함되어 있다. 내부 그래프가 아닌 경우에는 하나의 Figure는 하나의 윈도우를 가진다.
주피터 노트북에서는 윈도우객체가 생성되지 않지만 파이썬을 독립적으로 실행할 경우에는 하나의
Figure당 하나의 아이디와 윈도우를 별도를 가진다

 원래는 Figure를 생성하려면 figure라는 명령을 사용해서 Figure객체를 얻어야 하지만 일반적으로 plot명령등을 실행하면 자동으로 Figure를 생성해 주기 때문에 figure명령을 잘 사용하지는 않는다. figure명령을 명시적으로 사용하는 경우에` 여러개의 윈도우를 동시`에 띄우거나 `주피터 노트북에서 그림의 크기를 설정`할 경우 사용한다.

```python
# 그림크기 설정하기

np.random.seed(0)
f1 = plt.figure(figsize = (10,3))

plt.title('그림크기설정하기')
plt.plot(np.random.randn(100)) # 정규분포범위
plt.show()
```

##### Axes객체와 subplot명령

하나의 윈두우(Figure)안에 여러개의 그래프를 배열형태로 보여야 할 경우에 Figure안에 있는각각의 그래프는 `Axes라고 불리는 객체`이다.

Figure안에 Axes를 생성하려면 `subplot명령`을 사용해서 명시적으로 Axes객체를 얻어야 한다. 그러나 plot명령을 바로 사용해도 자동으로 Axes객체를 생성해 준다.
subplot명령은 그리드(grid)형태의 Axes객체를 생성하는데 `Figure가 행렬(Matrix)`이고 `Axes가 행렬의 원소`라고 생각하면 된다. 주의할 점은 첫번째 `Axes객체를 가리키는 숫자는 0이 아니라 1인 것에 주의` 해야 한다.

<img src ='./images/24.데이터분석_matplotlib_01.png' height = 400 width = 400>
  

```python
# 1. subplot() subplot(221) = subplot(2,2,1)
# 두개의 그래프를 그릴 경우 행이 2이고 열이 1인 행렬이다. subplot
# 명령은 한개의 figure에 공간을 나누어 여러 그래프를 넣을 시 유용하다.
# subplot(x,y,z) 인수를 가진다. subplot(행, 열, Axes객체위치)
x1 = np.linspace(0.,5.)
x2 = np.linspace(0.,2.)
y1 = np.cos(2*np.pi*x1) * np.exp(-x1)
y2 = np.sin(2*np.pi*x2) 

# Axes 객체 생성
ax1 = plt.subplot(2,1,1)
plt.plot(x1,y1,'yo-')

ax2 = plt.subplot(2,1,2)
plt.plot(x2,y2,'b^:')

plt.tight_layout() # 그래프의 간격을 자동으로 맞춤
plt.show()
```

```python
#2. subplot()
np.random.seed(0)

plt.subplot(2,2,1)
plt.plot(np.random.rand(5))
plt.title('첫번째')

plt.subplot(2,2,2)
plt.plot(np.random.rand(5))
plt.title('두번째')

plt.subplot(2,2,3)
plt.plot(np.random.rand(5))
plt.title('세번째')

plt.subplot(2,2,4)
plt.plot(np.random.rand(5))
plt.title('네번째')

plt.tight_layout() # 그래프의 간격을 자동으로 맞춤
plt.show()
```

```python
# 3. subplot() : 그룹별
names = ['GroupA', 'GruopB', 'GroupC', 'GroupD', 'GroupE','GroupF']
values = [5,10,8,9,7,5]

plt.figure(1, figsize = (10,3))

plt.subplot(131)
plt.bar(names, values)
plt.title('Bar')

plt.subplot(132)
plt.scatter(names, values)
plt.title('Scatter')

plt.subplot(133)
plt.plot(names, values)
plt.title('Line')

plt.suptitle('전체 제목')
plt.tight_layout()
plt.show()
```

```python
# Stacked Bar 그래프

N = 5
men_means = (25,30,35,40,27)
women_means = (21,25,32,32,23)
men_std = (2,3,4,1,2)
women_std = (3,5,2,3,3)

ind = np.arange(N)
width = 0.35
p1 = plt.bar(ind, men_means, width, yerr=men_std)
p2 = plt.bar(ind, women_means, width, yerr=women_std, bottom=men_means)

plt.ylabel('정수')
plt.xlabel('성별 그룹별 점수')
plt.xticks(ind,('G1','G2','G3','G4','G5'))
plt.yticks(np.arange(0,81,10))
plt.legend((p1[0],p2[0]),('남','녀'))

plt.show()

```

```python
# 이노데이션
# 그래프에 화살표를 그린 후 그 화살표에 문자열을 출력하는 기능이다.
# 예를 '이 값이 최소값' 이런식으로 화살표를 그려서 표현하고자 할 때 사용하는 기능
# plt.annotation('문자열',xy, xytext, arrowprops)식으로 사용한다.
# 문자열 : 어노테이션나타날 문자열
# xy : 화살표가 가리키는 점의 위치
# xytext : 문자열이 출력될 위치
# arrowprops : 화살표의 속성으로 컬러등을 정의한다.

x= np.arange(1,10)
y = x*5
plt.plot(x,y)
plt.annotate('이 곳이 최소값',xy=(2,10),xytext=(5,20),arrowprops={'color':'red'})
plt.show()
```

```python
# Gradient descent
# 경사하강법
# 특정 위치에 대한 스칼라함수가 존재할 경우 그 위치의 주변에서 가장 함수의 값이 커지는 방향
# (최적의값)을 가르키는 벡터이다.
# 벡터는 방향과 크기가 있는 값이라면 스칼라는 값만 존재한다. 
# 여기서 말하는 벡터는 화살표를 의미하며 gradient의 기울기 값이다.
# f(x) = x^2의 gradient를 그리기'

x = np.linspace(-2,3,100)
y = x**2
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(x,y, color ='k', linestyle='-', lw=3)
ax.scatter(2,4,color='y', s=500, alpha=0.5)
ax.arrow(2,4,0.4,1.6,lw=3,head_width=0.2, head_length=0.1, fc= 'r', ec = 'r')
ax.arrow(2,4,-0.4,-1.6,lw=3,head_width=0.2, head_length=0.1, fc= 'b', ec = 'b')
ax.scatter(1.6,2.4,color='y', s=500, alpha=0.5)
ax.arrow(1.6,2.4,-0.4,-0.94,lw=3,head_width=0.2, head_length=0.1, fc= 'b', ec = 'b')

plt.show()
```

##### 연습문제 3. 영화평점 분석결과 시각화(자유형)

```python
import pandas as pd
df_movies = pd.read_csv('../data/movilens/movies.csv',delimiter = ',')
df_ratings = pd.read_csv('../data/movilens/ratings.csv',delimiter = ',')
```

```python
mean_df = df_ratings.groupby('movieId')['rating'].mean()
df = pd.merge(mean_df, df_movies, left_on='movieId', right_on='movieId', how='inner')


# df.iloc[:,0] movieid  1 title 2 rating

# 카테고리 추출 - 총 20
a = df['genres'].str.split('|')
b = []
len(a)

for i in range(len(a)):
    for j in range(len(a[i])):
        b.append(a[i][j])
c = set(b)
c = list(c)
c[1] = 'Romance'



# 전체 영화의 평균점수 분포
fig, axs = plt.subplots(20,1,figsize=(5,15))
axs[0].hist(df_ratings.groupby('movieId')['rating'].mean())
for i in range(1,len(c)):
    axs[i].hist(df[df['genres'].str.contains(c[i])].iloc[:,1], label=c[i])
    

plt.suptitle('영화 평점 분포도')

plt.show()
```
